```python
banco1['date'] = pd.to_datetime(banco1["date"], format='%d-%B-%Y')
banco1.head()
```

```python
banco1['contact_year'] = banco1['date'].dt.year
banco1.head()
```

```python
banco1['contact_month'] = banco1['date'].dt.month
banco1.head()
```

```python
banco1['contact_month'] = banco1['contact_month'].fillna(0).astype(int)
banco1['contact_year'] = banco1['contact_year'].fillna(0).astype(int)
banco1.head()
```

```python
banco1['contact_year'].unique()
```

```python
banco1 = banco1.rename(columns={'id_':'ID'})
banco1.head()
```

```python
banco1.drop('Unnamed: 0', axis=1, inplace=True)
banco1.head()
```

```python
banco1.drop('latitude', axis=1, inplace=True)
banco1.drop('longitude', axis=1, inplace=True)
banco1.head()
```

```python
banco1 = banco1.rename(columns={'y':'product'})
banco1.head()
```

```python
banco1['default'] = banco1['default'].apply(lambda x: 'yes' if x == 1 else 'no')
banco1.head()
```

```python
banco1['housing'] = banco1['housing'].apply(lambda x: 'yes' if x == 1 else 'no')
banco1.head()
```
